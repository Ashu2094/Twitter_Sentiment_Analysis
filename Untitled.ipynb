# Cleaning The Text

```python
import string
string.punctuation
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords.words('english')
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def  tweet_cleaning(message):
    punc_removed = [char for char in message if char not in string.punctuation]
    punc_removed_join = ''.join(punc_removed)
    punc_removed_join_clean = [word for word in punc_removed_join.split() if word.lower() not in stopwords.words('english')]
    return punc_removed_join_clean
```

```python
tweets_df_cleaned = new_tweets_df['tweet'].apply(tweet_cleaning)
```

```python
# print(tweets_df_cleaned[5])
```

```python
vectorizer = CountVectorizer(analyzer = tweet_cleaning)
tweet_countvectorizer = CountVectorizer(analyzer=tweet_cleaning, dtype='uint8').fit_transform(new_tweets_df['tweet']).toarray()
```

```python
tweet_countvectorizer.shape
```

```python
X = tweet_countvectorizer
```

```python
Y = new_tweets_df['label']
```

# Creating A Naive Bayes Classifier

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
NB_classifier = MultinomialNB()
NB_classifier.fit(X_train,Y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
Y_Predict_Test = NB_classifier.predict(X_test)
```

```python
cm = confusion_matrix(Y_test,Y_Predict_Test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(Y_test,Y_Predict_Test))
```
